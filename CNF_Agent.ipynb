# CNF Agent 

This agent's knowledgebase is a conjunction of disjunctions of literals.  The literals take the form M(i, j).  THis means there is a mine at (i, j).  Its negation is that (i. j) is safe.  

The agent has a .kb attribute that represents its knowledgebase.  It is a list of clauses that are assumed to be AND'ed together.  Each clause is a list of literals that are assumed to be OR'ed together.

This agent will go through the following functions to solve the board: 

    (1) Discover single literal clauses to deduce mine and safe locations 
        + If a single literal is M(i, j), then (i, j) is a mine 
        + If a single literal is -M(i, j), then (i, j) is not a mine 


    (2) Query the knowledgebase for -M(i, j).  
         If this query is unsat, then (i, j) is a mine 

    (3)  Query the knowledgebase for M(i, j). 
         If this query is unsat, then (i, j) is safe 

    (4)  Uncover random covered, unflagged cell
