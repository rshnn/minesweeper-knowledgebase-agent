# Bonus 

A new board object was created called JerkBoard.  The JerkBoard takes in a parameter called the `fog_probability`.  This is the chance that the board will not notify the agent of the hint value when uncovering safe cells.    

A modified CNF Agent was designed to cope with this jerk board.
