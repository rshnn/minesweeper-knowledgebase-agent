# Basic Agent


The basic agent solves the minesweeper game using local search across the maze.  After uncovering one cell, it will iterate through all the cells on the board to determine if it can make any judgements about mine positions and safe cells.  If it cannot make any judgements given the state of the board, it will randomly chose a covered cell to uncover.
